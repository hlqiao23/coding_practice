### Input:

```python
Order:
{
    "country": "US", // or "CA" for the CA order
    "items": [
        {"product": "mouse", "quantity": 20},
        {"product": "laptop", "quantity": 5}
    ]
}

Shipping Cost:
{
    "US": [
        {"product": "mouse", "cost": 550},
        {"product": "laptop", "cost": 1000}
    ]
    "CA": [
        {"product": "mouse", "cost": 750},
        {"product": "laptop", "cost": 1100}
    ]
}
```

### Part 1:
Given the orders and shipping costs, calculate the order shipping cost. 
Example:
```python
calculate_shipping_cost(order_us, shipping_cost) == 16000
calculate_shipping_cost(order_ca, shipping_cost) == 20500
```


````python
def calculate_shipping_cost_part1(order, shipping_cost):


    cty = order['country']

    total_cost = 0

    for record in order['items']:
        product = record['product']
        quantity = record['quantity']

        for s in shipping_cost[cty]:
            if s['product'] == product:
                total_cost += s['cost'] * quantity

    return total_cost
````

````python
order_us = {
    "country": "US", 
    "items": [
        {"product": "mouse", "quantity": 20},
        {"product": "laptop", "quantity": 5}
        ]
}
order_ca = {
    "country": "CA",
    "items": [
        {"product": "mouse", "quantity": 20},
        {"product": "laptop", "quantity": 5}
        ]
}

shipping_cost = {
    "US": [
        {"product": "mouse", "cost": 550}, 
        {"product": "laptop", "cost": 1000}, 
    ], 
    "CA": [
        {"product": "mouse", "cost": 750}, 
        {"product": "laptop", "cost": 1100}, 
    ], 
}

print(calculate_shipping_cost_part1(order_us, shipping_cost))
print(calculate_shipping_cost_part1(order_ca, shipping_cost))
````

````python
def get_total_cost(price_range, quantity):
    total = 0
    while quantity > 0:
        max, price = price_range.pop(0)
        # determine the quantity to calculate in this iteration
        if max is None:
            quantity_now = quantity
        else:
            quantity_now = min(max, quantity)
        total += quantity_now * price
        # substract the calculated quantity
        quantity -= quantity_now
    return total

def calculate_shipping_cost_part2(order, shipping_cost):

    cty = order['country']

    total_cost = 0

    # loop through all single orders
    for record in order['items']:
        product = record['product']
        quantity = record['quantity']
        # print(product, quantity)
        
        # for each product, quantity, find the shipping cost from the mapping
        for s in shipping_cost[cty]:
            # find the product
            if s['product'] == product:
                # convert the dictionary of costs into a list of tuples (queue)
                price_range = []
                for c in s['costs']:
                    price_range.append((c['maxQuantity'], c['cost']))
                total_cost +=get_total_cost(price_range, quantity)
    
    return total_cost
````

### Part 3:
The shipping cost has added one more attribute, either 'fixed' or 'incremental'. 
fixed: no matter what quantity it is, the total price is fixed as the cost. 
incremental: the total price will be the quantity fall into this bracket * price per item.

````python
shipping_cost ={
    "US": [
        {"product": "mouse",
         "costs": [{"type": "incremental", "minQuantity": 0, "maxQuantity": None, "cost": 550}], 
        },
        {"product": "laptop",
         "costs": [{"type": "fixed", "minQuantity": 0,"maxQuantity": 2,"cost": 1000},
                   {"type": "incremental", "minQuantity": 3,"maxQuantity": None,"cost": 900}]
        }
    ],
    "CA": [
        {"product": "mouse",
         "costs": [{"type": "incremental", "minQuantity": 0, "maxQuantity": None, "cost": 750}], 
        },
        {"product": "laptop",
         "costs": [{"type": "fixed", "minQuantity": 0,"maxQuantity": 2,"cost": 1100},
                   {"type": "incremental", "minQuantity": 3,"maxQuantity": None,"cost": 1000}]
        }
    ]
}
````

````python
def get_total_cost(price_range, quantity):
    total = 0
    while quantity > 0:
        type, max, price = price_range.pop(0)

        # determine the quantity to calculate in this iteration
        if max is None:
            quantity_now = quantity
        else:
            quantity_now = min(max, quantity)

        # based on the type, calculate the total 
        if type == 'fixed':
            total += price
        else:   
            total += quantity_now * price
        # substract the calculated quantity
        quantity -= quantity_now
    return total

def calculate_shipping_cost_part3(order, shipping_cost):

    cty = order['country']

    total_cost = 0

    # loop through all single orders
    for record in order['items']:
        product = record['product']
        quantity = record['quantity']
        # print(product, quantity)
        
        # for each product, quantity, find the shipping cost from the mapping
        for s in shipping_cost[cty]:
            # find the product
            if s['product'] == product:
                # convert the dictionary of costs into a list of tuples (queue)
                price_range = []
                for c in s['costs']:
                    price_range.append((c['type'], c['maxQuantity'], c['cost']))
                total_cost +=get_total_cost(price_range, quantity)
    
    return total_cost
````

````python
print(calculate_shipping_cost_part3(order_us, shipping_cost))
print(calculate_shipping_cost_part3(order_ca, shipping_cost))
````
